```python
from IPython.display import display, HTML
import pandas as pd

visits = pd.read_csv('visits.csv',
                     parse_dates=[1])
cart = pd.read_csv('cart.csv',
                   parse_dates=[1])
checkout = pd.read_csv('checkout.csv',
                       parse_dates=[1])
purchase = pd.read_csv('purchase.csv',
                       parse_dates=[1])
# display(HTML(purchase.head(11).to_html()))
# print(visits.head(5))
# print(cart.head(5))
# print(checkout.head(5))
# print(purchase.head(5))
visit_cart_merge = pd.merge(visits, cart, how="left")
# print(visit_cart_merge.head(10))
total_visits = len(visit_cart_merge)
visitcart_null = visit_cart_merge[visit_cart_merge.cart_time.isnull()]

#What percent of users who visited Cool T-Shirts Inc. ended up not placing a t-shirt in their cart?
didnt_add_to_cart_percentage = float(len(visitcart_null) / float(total_visits))
print("{} percent of users who visited the page did not add a t-shirt to their cart".format(round(didnt_add_to_cart_percentage*100, 2)))

#Repeat the left merge for cart and checkout and count null values. What percentage of users put items in their cart, but did not proceed to checkout?
cart_checkout_merge = pd.merge(cart, checkout, how="left")
# print(cart_checkout_merge.head(10))
count_cart_checkout_null = cart_checkout_merge[cart_checkout_merge.checkout_time.isnull()]
checkout_didnt_buy_percent = float(len(count_cart_checkout_null)) / (len(cart_checkout_merge))
print("{} percent of users who made it to checkout  did not purchase a shirt".format(round(checkout_didnt_buy_percent*100, 2)))
# Merge all four steps of the funnel, in order, using a series of left merges. Save the results to the variable all_data.
all_data = visits.merge(cart, how='left').merge(checkout,how='left').merge(purchase, how='left')
# print(all_data.head())
all_data["initial_visit_to_final_purchase"] = all_data.purchase_time - all_data.visit_time
ave_time_checkout = [all_data.initial_visit_to_final_purchase.mean()]
print(ave_time_checkout)
display(HTML(all_data.head(55).to_html()))
```
