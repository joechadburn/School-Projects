```python
import pandas as pd
import numpy as np
from IPython.display import HTML

db = pd.read_csv('diabetes.csv')
display(db.head())
print(len(db.columns))
print(db.isnull().sum())
db.info()
db.describe()

db[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = db[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.nan)
db.describe()
print(db.isnull().sum())
display(db[db.isnull().any(axis=1)].head(50))
print(db['Outcome'].unique())
db['Outcome'] = db['Outcome'].replace('O', 0)
db['Outcome'] = db['Outcome'].astype('int64')
print(db.Outcome.unique())
db.info()
db.value_counts()
```
